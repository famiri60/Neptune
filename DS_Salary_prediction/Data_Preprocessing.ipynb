```python
def shorten_categories(categories,col, cutoff):
    #categorical_map = {}
    for i in range(len(categories)):
        if categories[i] >= cutoff:
            col[i] = 'PhD'
        else:
            col[i] = 'Master'
    return col
```

```python
degree_map = shorten_categories(data['Avg Salary(K)'].values, data['Degree'].values, 100.0)
data['Degree'] = degree_map
data['Degree'].value_counts()
```

```python
##[reference:https://www.dataquest.io/blog/tutorial-add-column-pandas-dataframe-based-on-if-else-condition/]
# create a list of our conditions
conditions = [
    (data['Revenue']=='Less than $1 million (USD)') | (data['Revenue']=='$1 to $5 million (USD)') | (data['Revenue']=='$5 to $10 million (USD)') | (data['Revenue']=='$10 to $25 million (USD)') | (data['Revenue']=='$25 to $50 million (USD)'),
    (data['Revenue']=='$50 to $100 million (USD)') | (data['Revenue']=='$100 to $500 million (USD)') | (data['Revenue']=='$500 million to $1 billion (USD)'),
    (data['Revenue']=='$1 to $2 billion (USD)') | (data['Revenue']=='$2 to $5 billion (USD)') | (data['Revenue']=='$5 to $10 billion (USD)') | (data['Revenue']=='$10+ billion (USD)'),
    data['Revenue']=='Unknown / Non-Applicable'
    ]
values = ['Small', 'Medium', 'Large', 'na']
data['Company_Size'] = np.select(conditions, values)
n=len(data[data['Company_Size']=='na'])
```

```python
rev=data['Company_Size'].copy().values
#n=len(data['Revenue'])
rev1=data['Size'].copy().values
for j in range(len(data['Company_Size'])):
    if rev[j]=='na':
        if (rev1[j]=='1 - 50 ') | (rev1[j]=='51 - 200 ') | (rev1[j]=='201 - 500 '):
            rev[j]='Small'
        if (rev1[j]=='1001 - 5000 ') | (rev1[j]=='501 - 1000 '):
            rev[j]='Medium'
        if (rev1[j]=='5001 - 10000 ') | (rev1[j]=='10000+ '):
            rev[j]='Large'

data['Company_Size']=rev
data=data.drop(data[data['Company_Size']=='na'].index)
#len(data[data['Company_Size']=='Large'])
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

max_depth = [None, 20,40,60,80,100,200]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs_reg0=gs.fit(X0_train, y0_train.values)

```
