```python
import chardet
import pandas as pd

# Function to detect the encoding
def detect_encoding(file_path):
    rawdata = open(file_path, 'rb').read()
    result = chardet.detect(rawdata)
    encoding = result['encoding']
    return encoding

# File path
file_path = './data/oracle.csv'

# Detect encoding
encoding = detect_encoding(file_path)

# Read the file using the detected encoding
df = pd.read_csv(file_path, encoding=encoding)

# Perform data operations if needed

# Save the modified DataFrame back to a CSV file with UTF-8 encoding
df.to_csv('oracle.csv', encoding='utf-8', index=False)
```

```python
import pandas as pd
import os

# Set the directory where your CSV files are located
csv_directory = './data'

# Initialize an empty list to store DataFrames
dataframes = []

# List CSV files in the directory
csv_files = [f for f in os.listdir(csv_directory) if f.endswith('.csv')]

# Loop through the CSV files
for file in csv_files:
    # Extract the company name from the file name (assuming the format is 'company_name.csv')
    company_name = file.replace('.csv', '')
    
    # Load the CSV file into a DataFrame
    df = pd.read_csv(os.path.join(csv_directory, file))

    
    # Add a new column 'Company' with the company name
    df['Company'] = company_name
    
    # Append the DataFrame to the list
    dataframes.append(df)

# Concatenate all DataFrames into one
combined_data = pd.concat(dataframes, ignore_index=True)

# You now have a single DataFrame with all your data

```

```python
combined_data.info()
```

```python
# Convert 'Latitude' and 'Longitude' columns to numeric (float) data types
combined_data['Latitude'] = pd.to_numeric(combined_data['Latitude'], errors='coerce')
combined_data['Longitude'] = pd.to_numeric(combined_data['Longitude'], errors='coerce')
```

```python
# Define a function to create the 'Location' column
def create_location(row):
    if pd.notna(row['State']):
        return f"{row['City']}, {row['State']}, {row['Country']}"
    else:
        return f"{row['City']}, {row['Country']}"

# Apply the function to create the 'Location' column
combined_data['Location'] = combined_data.apply(create_location, axis=1)

combined_data
```

```python
# You can save it to a new CSV file if needed
combined_data.to_csv('combined_data.csv', index=False)
```

```python
import folium
import os

# Define a folder to save the map images
image_folder = 'map_images'

def generate_map_image(row):
    if row['Latitude'] != '-' and row['Longitude'] != '-':
        location = (float(row['Latitude']), float(row['Longitude']))
        m = folium.Map(location=location, zoom_start=15)
        folium.Marker(location=location, popup=row['Location']).add_to(m)
        
        # Generate a unique image file name based on the index or any other unique identifier
        image_filename = f'{row.name}.html'  # You can change the naming logic here
        
        # Save the map image to the image folder
        image_path = os.path.join(image_folder, image_filename)
        m.save(image_path)
        
        return image_filename
    else:
        return 'Invalid Location'

# Create a new column with image filenames
combined_data['Map Image'] = combined_data.apply(generate_map_image, axis=1)



# Save the modified DataFrame to a new CSV
combined_data.to_csv('combined_data.csv', index=False)

```

```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage you want to scrape
url = 'https://en.wikipedia.org/wiki/Google_data_centers'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all the tables on the webpage
    tables = soup.find_all('table')

    # Loop through each table and extract the data
    for table in tables:
        # Extract the rows from the table
        rows = table.find_all('tr')
        for row in rows:
            # Extract the cells from each row
            cells = row.find_all(['td', 'th'])
            row_data = [cell.get_text(strip=True) for cell in cells]
            print(row_data)
else:
    print(f'Failed to retrieve the webpage. Status code: {response.status_code}')
```
